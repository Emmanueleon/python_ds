```python
## Importar la librerìa
import requests

## Guardar la url a descargar
URL = 'https://tripleten-com.github.io/simple-shop_es/'

## Usar el método `.get` para bajar la estructura de la página web
req = requests.get(URL)

print(req.text)

# Usar el método `.status_code` para obtener el código de error
print(req.status_code)
```

Métodos más utilizados en expresiones regulares usando la libería `re`. 
1. `search()`. Busca un patrón en una cadena, devolviendo solo la primera subcadena que encuentra.
2. `split()`. Divide una cadena en los puntos donde aparece el patrón.
3. `sub()`. Busca el patrón de subcadena dentro de una cadena y lo reemplaza con la subcadena repl (replace)
4. `findall()`. Devuelve una lista de todas las subcadenas en una cadena que coinciden con el patrón. Es como el método `search()`, pero devuelve todas las búsquedas.
